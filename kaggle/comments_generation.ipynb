```python
import openai
def response(prefix, codex_name="text-davinci-003", 
                     max_tokens=1024,
                     temperature=0.0,
                     top_p=1.0):

    key = "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"
    try:
        response = openai.Completion.create(
                        engine=codex_name,
                        prompt=prefix,
                        suffix=None,
                        temperature=temperature,
                        max_tokens=max_tokens,
                        api_key=key,
                        top_p=top_p,
                        frequency_penalty=0,
                        presence_penalty=0,
                        stop=["</code>", "# SOLUTION END"],
                        logprobs=1,
                        n=1,
                    )
    except (openai.error.RateLimitError, openai.error.APIConnectionError) as e:
            print(type(e), e)
            return 'Please check the Error!!!'
    return response['choices'][0]['text']
```

```python
import os
root_path = 'kernels_train'
kernel_path = 'ilyai332/unemployment-in-america-per-us-state/unemployment-in-america-per-us-state.ipynb'
kernel_path = os.path.join(root_path, kernel_path)
import json
nb_code = json.load(open(kernel_path))
data = nb_code['cells']
#data
```

```python
i = 0
while i < len(data):
    cell = data[i]
    code = ''
    prompt = '''
The code:
'''
    # get cell code
    if cell['cell_type'] == 'code':
        for s in cell['source']:
            code += s
    else:
        i = i + 1
        continue
    if len(code) == 0:
        print('ERROR: No code found !!!')
        i = i + 1
        continue
    prompt += code
    print(prompt)
    # get comment from chatgpt
    prefix = prompt + ans
    out = response(prefix)
    out = out.replace('This code', '').replace('The code', '').replace('This function', '').strip()
    print('>>', out)
    # insert a cell comment
    new_comment_cell = {
               "cell_type": "markdown",
               #"execution_count": None,
               "metadata": {},
               "outputs": [],
               "source": ['- ' + out]
              }
    data.insert(i, new_comment_cell)
    i = i + 2
    
with open(kernel_path, 'w', encoding='UTF-8') as f:
    json.dump(nb_code, f)
```
